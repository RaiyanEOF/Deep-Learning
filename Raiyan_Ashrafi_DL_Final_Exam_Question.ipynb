# Question 01: [ Marks 10 ]

## Explain the concept of backpropagation in detail. Describe how the error is propagated from the output layer to the hidden layers. Derive the mathematical expressions for weight updates using the chain rule, and explain how learning rate affects the training process.


# Write Answer 01:

**Backpropagation is the method a neural network uses to learn from its mistakes by sending the error backward from the output layer to the hidden layers. First, the network makes a prediction in the forward pass, then it compares that prediction with the correct answer to find the error. In backpropagation, this error is passed backward layer by layer so each weight knows how much it contributed to the mistake. Mathematically, this is done using the chain rule, which helps find the gradient of the loss with respect to each weight: the update rule is usually written as weight new = weight old − learning rate × gradient. The gradient tells us the direction and size of the change needed to reduce error. The learning rate controls how big each update is: a very small learning rate makes training slow but stable, while a very large one can make training unstable or cause it to miss the best solution. So, backpropagation helps the network improve step by step by adjusting weights in the right direction, and the learning rate decides how fast those adjustments happen.**

# Backpropagation for a Neural Network with One Hidden Layer (Sigmoid)

## Forward Pass

Let the input be `x`.

Hidden layer:
z1 = W1 · x + b1  
a1 = sigmoid(z1)

Sigmoid function:
sigmoid(z) = 1 / (1 + e^(-z))

Output layer:
z2 = W2 · a1 + b2  
y_hat = a2 = sigmoid(z2)

---

## Loss Function

L = 1/2 (y - y_hat)^2

---

## Backward Pass

Step 1: Error at output layer

dL/dy_hat = (y_hat - y)

Since y_hat = sigmoid(z2),

sigmoid'(z) = sigmoid(z) (1 - sigmoid(z))

delta2 = dL/dz2  
delta2 = (y_hat - y) * y_hat * (1 - y_hat)

---

Step 2: Gradients for output layer

dL/dW2 = delta2 · (a1)^T  
dL/db2 = delta2

---

Step 3: Propagate error to hidden layer

delta1 = (W2^T · delta2) * a1 * (1 - a1)

---

Step 4: Gradients for hidden layer

dL/dW1 = delta1 · (x)^T  
dL/db1 = delta1

---

## Weight and Bias Update

W2_new = W2_old - learning_rate * dL/dW2  
b2_new = b2_old - learning_rate * dL/db2

W1_new = W1_old - learning_rate * dL/dW1  
b1_new = b1_old - learning_rate * dL/db1

# Question 03: [Marks 5]

## Discuss the vanishing gradient problems in deep neural networks. Explain why they occur during backpropagation, how activation functions contribute to this issue, and suggest at least three techniques to overcome these problems.


## Write Answer 03:

**The vanishing gradient problem happens in deep neural networks when the gradients become extremely small as they are propagated backward from the output layer to the earlier hidden layers during backpropagation. This occurs because, in backpropagation, gradients are calculated using the chain rule, which multiplies many small derivative values together. When activation functions like sigmoid or tanh are used, their derivatives are always less than 1, so multiplying them repeatedly across many layers makes the gradient shrink exponentially. As a result, the weights in the earlier layers receive almost no updates, causing the network to learn very slowly or not at all. Activation functions contribute to this issue because sigmoid and tanh saturate for large positive or negative inputs, where their gradients become very close to zero. To overcome this problem, several techniques are used: (1) using activation functions like ReLU that do not saturate in the positive region and have larger gradients, (2) using batch normalization to normalize layer inputs and stabilize gradient flow, and (3) using architectures like residual networks (skip connections) that allow gradients to pass more directly to earlier layers.**

## Describe the architecture of an Artificial Neural Network (ANN) in detail. Explain the role of each component (neurons, weights, bias, activation functions). Also discuss how increasing the number of hidden layers and neurons affects model performance and overfitting.


## Write Answer 04:

**An Artificial Neural Network (ANN) is a computing model inspired by the human brain, made up of layers of interconnected units called neurons organized as an input layer, one or more hidden layers, and an output layer. Each neuron receives inputs, multiplies them by weights, adds a bias, and then passes the result through an activation function to produce an output. The activation function (such as sigmoid, ReLU, or tanh) introduces non-linearity so the network can learn complex patterns rather than just simple linear relationships. During training, the network adjusts the weights and biases to minimize error and improve predictions. Increasing the number of hidden layers and the number of neurons allows the model to learn more complex features and patterns, which can improve performance on difficult tasks. However, if the network becomes too large compared to the amount of training data, it may start memorizing the data instead of learning general patterns, leading to overfitting. Therefore, a balance must be maintained between model complexity and generalization ability.**

# Question 05: [ Marks 10 ]

## Compare Artificial Neural Networks (ANN) and Convolutional Neural Networks (CNN) in detail. Discuss their architectures, working principles, advantages, limitations, and typical use cases such as image processing, tabular data, and pattern recognition.

## Write Answer 05:

**Artificial Neural Networks (ANN) and Convolutional Neural Networks (CNN) are both deep learning models made of layers of neurons, but they differ in architecture and how they process data. A basic ANN connects every neuron in one layer to every neuron in the next layer, treats input as a flat vector, and learns patterns through weighted sums, biases, and activation functions; it works well for tabular data, simple pattern recognition, and problems where spatial structure is not important, but it becomes inefficient and parameter-heavy for images because it ignores the 2D structure and requires many weights. In contrast, a CNN is specially designed for grid-like data such as images: it uses convolutional layers with small filters (kernels) that slide over the input to detect local features (edges, textures), pooling layers to reduce size and retain important information, and fewer parameters due to weight sharing, which makes it more efficient and better at capturing spatial patterns. CNNs automatically learn hierarchical features (from simple edges to complex objects) and are widely used in image processing, video analysis, and pattern recognition tasks. ANNs are easier to design and faster for small, structured datasets but struggle with high-dimensional inputs like images, while CNNs excel at visual data but are more complex and computationally intensive. Thus, ANNs are preferred for tabular or numerical data, whereas CNNs are preferred for image and spatial pattern tasks.**

Activation Functions in Neural Networks

Activation functions add non-linearity so a neural network can learn complex patterns instead of only linear relationships.

1) Sigmoid

Formula:
sigmoid(x) = 1 / (1 + e^(-x))

Output range:
0 to 1

Advantages:
- Smooth and differentiable
- Output can be treated as probability
- Useful in binary classification output layer

Disadvantages:
- Vanishing gradient for very large or very small values
- Not zero-centered, which can slow learning

Suitable use:
- Output layer for binary classification problems

---

2) Tanh (Hyperbolic Tangent)

Formula:
tanh(x) = (e^x - e^(-x)) / (e^x + e^(-x))

Output range:
-1 to 1

Advantages:
- Zero-centered, faster learning than sigmoid
- Smooth and differentiable

Disadvantages:
- Still suffers from vanishing gradient for large inputs

Suitable use:
- Hidden layers in small or shallow networks

---

3) ReLU (Rectified Linear Unit)

Formula:
ReLU(x) = max(0, x)

Output range:
0 to infinity

Advantages:
- Very simple and fast to compute
- Reduces vanishing gradient problem
- Creates sparse activation (many zeros), which is efficient

Disadvantages:
- Dying ReLU problem (neurons may get stuck at zero)

Suitable use:
- Hidden layers in deep neural networks

---

Why ReLU is widely used in deep learning

ReLU is computationally efficient, helps avoid the vanishing gradient problem, and allows deep networks to learn faster and more effectively. That is why it is the default activation function in most deep learning models.

Step 1: Apply padding

Because padding is 1, we add a border of zeros around the input.

## Write Answer 08:

Pooling in CNN

Pooling is an operation used in Convolutional Neural Networks after the convolution layer. It reduces the size of the feature map while keeping the most important information. This helps make the network faster and prevents overfitting.

Pooling works by taking a small window (for example 2×2) and moving it across the feature map. From each window, a single value is selected to create a smaller output.

---

**1.** MaxPooling

Operation:
From each window, the maximum value is selected.

Example (2×2 window):

[ 1  5

  2  3 ]

Max value = 5

Advantages:
- Keeps the most important and strongest features
- Helps detect edges and important patterns clearly
- Works well for image recognition

Impact on feature extraction:
MaxPooling preserves the most dominant features from the feature map.

---

**2.** AveragePooling

Operation:
From each window, the average of all values is taken.

Example (2×2 window):

[ 1  5

  2   3 ]

Average = (1 + 5 + 2 + 3) / 4 = 2.75

Advantages:
- Keeps general information from the region
- Produces smoother feature maps

Impact on feature extraction:
AveragePooling keeps overall information but may lose strong features.

---

Comparison

MaxPooling:
- Selects maximum value
- Highlights strong features
- Commonly used in CNNs

AveragePooling:
- Selects average value
- Smooths features
- Less commonly used for images

---

How pooling reduces overfitting and computation

Pooling reduces the size of the feature map. This means:
- Fewer parameters to learn
- Less computation required
- Lower chance of overfitting because the network becomes simpler

# Question 10: [ Marks 20 ]

## Design and implement a deep Convolutional Neural Network (CNN) for a multi-class image classification problem using PyTorch. The network should consist of four convolutional blocks, where each block includes two convolution layers followed by activation and normalization, and a pooling layer at the end. The number of filters should increase progressively across blocks. After the convolutional part, flatten the feature maps and connect them to fully connected layers with dropout for regularization, and produce final outputs corresponding to K classes. Also explain the role of each part of the network in detail.

# Write Answer 10:

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
class DeepCNN(nn.Module):
    def __init__(self, K):
        super(DeepCNN, self).__init__()
        self.block1 = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.Conv2d(32, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.block2 = nn.Sequential(
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.block3 = nn.Sequential(
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.block4 = nn.Sequential(
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.fc1 = nn.Linear(256*8*8,512)
        self.dropout = nn.Dropout(0.5)
        self.fc2 = nn.Linear(512,K)
    def forward(self,x):
        x = self.block1(x)
        x = self.block2(x)
        x = self.block3(x)
        x = self.block4(x)
        x = x.view(x.size(0),-1)
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x
```

## Explanation

Explanation of the Deep CNN Architecture

This CNN is designed for multi-class image classification. It contains four convolutional blocks followed by fully connected layers.

Each convolutional block has:
- Two convolution layers
- Activation function (ReLU)
- Batch Normalization
- A Max Pooling layer at the end

The number of filters increases in every block:
32 -> 64 -> 128 -> 256

This helps the network learn simple features first (edges, lines) and then more complex features (shapes, objects).

---

Block 1

Input image (for example 128×128×3) goes into the first block.

Two convolution layers extract basic features.  
Batch Normalization makes training stable.  
ReLU adds non-linearity.  
Max Pooling reduces the size to half.

Output becomes smaller in height and width but deeper in channels.

---

Block 2

Takes output from Block 1.  
Again two convolutions + ReLU + BatchNorm.  
Max Pooling reduces size further.

Now the network starts detecting more meaningful patterns.

---

Block 3

Same structure, but with more filters.  
Learns more complex features from the image.

---

Block 4

Final convolution block with highest number of filters.  
Captures very detailed and abstract features.

---

Flattening

After block 4, the feature map is converted into a single long vector.  
This is needed to connect with fully connected layers.

---

Fully Connected Layer (fc1)

This layer learns combinations of all extracted features.  
ReLU is used again for non-linearity.

---

Dropout

Randomly turns off some neurons during training.  
This helps prevent overfitting.

---

Final Layer (fc2)

Produces output for K classes.  
These outputs are used to decide which class the image belongs to.
