# **Write** Answer 01:
**For Question 1, the Iris dataset was chosen because it is a simple and widely used dataset for classification tasks. The dataset contains four numerical features and three classes, and it does not have missing values or categorical inputs that need encoding. The features were standardized to keep them on the same scale, which helps neural networks train more efficiently. After preprocessing, the input features (X) and target labels (y) were separated, then converted from NumPy arrays into PyTorch tensors so they could be used for deep learning. Finally, the dataset was split into training and testing sets using an 80/20 ratio, resulting in 120 samples for training and 30 samples for testing, which allows the model to learn from the training data and be evaluated on unseen test data.**




```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import numpy as np
import torch

iris = load_iris()
X = iris.data
y = iris.target
print("Dataset loaded:", X.shape[0], "samples,", X.shape[1], "features.")

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_np = np.array(X_scaled,dtype=np.float32)
y_np = np.array(y,dtype=np.longlong)

X_tensor = torch.tensor(X_np)
y_tensor = torch.tensor(y_np)

X_train_np,X_test_np,y_train_np,y_test_np = train_test_split(X_np,y_np,test_size = 0.2,random_state = 42)
X_train = torch.tensor(X_train_np)
X_test = torch.tensor(X_test_np)
y_train = torch.tensor(y_train_np)
y_test = torch.tensor(y_test_np)

```

## Write Answer 02:
**For Question 2, a neural network was designed with an input layer, one hidden layer, and an output layer. The input layer has 4 neurons because the Iris dataset contains four features. A hidden layer with around 10 neurons was chosen to provide enough capacity for the model to learn patterns without making the network too complex and causing overfitting. The ReLU activation function was used in the hidden layer because it is simple, computationally efficient, and helps the model learn non-linear relationships. The output layer contains 3 neurons to represent the three classes of the dataset, allowing the model to predict the correct class for each input sample.**

```python
import torch.nn as nn

class IrisNeuralNetwork(nn.Module):
  def __init__(self,input_dim = 4,hidden_dim = 10,output_dim = 3):
    super(IrisNeuralNetwork,self).__init__()
    self.fc1 = nn.Linear(4,10)
    self.relu = nn.ReLU()
    self.fc2 = nn.Linear(10,3)

  def forward(self,x):
    x = self.fc1(x)
    x = self.relu(x)
    x = self.fc2(x)
    return x

model = IrisNeuralNetwork()
print(model)
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

num_epochs = 100
loss_history = []

for epoch in range(1, num_epochs+1):

    outputs = model(X_train)
    loss = criterion(outputs,y_train)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    loss_history.append(loss.item())

    if epoch%10==0:
        print(f"Epoch {epoch}, Loss = {loss.item():.4f}")
```

# Training Analysis

Answer the following:

Why must gradients be reset every epoch?

What happens if learning rate is too high?

What happens if learning rate is too small?

Why do we define layers inside the constructor (__init__) and not inside forward()?


## Write Answer 07:
**Resetting Gradients: In PyTorch, gradients accumulate by default. When you call loss.backward(), gradients for each parameter are added to any existing values in grad. If we do not zero them, then gradients from multiple iterations combine, leading to incorrect updates. As explained in PyTorch documentation, failing to call optimizer zero_grad() will cause the optimizer to use stale, accumulated gradients from previous batches, preventing proper learning. Therefore, we reset zero gradients at the start of each iteration so that each update step uses only the current batch’s gradient.**
** **

**Learning Rate Too High: A too-large learning rate makes each update jump aggressively in parameter space. This can cause the training loss to oscillate or diverge instead of converging. As noted in convergence analysis a learning rate that is too high can cause the network to overshoot the optimal solution and diverge. In other words, the updates become so large that they miss the minimum and may even increase loss, preventing convergence.**
** **

**Learning Rate Too Low: Conversely, a very small learning rate makes the training progress extremely slow. The optimizer takes tiny steps, so many more epochs are needed. In some cases, it may get stuck in a suboptimal local minimum because it cannot move far enough to escape. A learning rate that is too low can cause the network to converge very slowly or get stuck in a local minimum.**
** **

**Layers in __init__ vs forward(): Pytorch expects learnable layers to be defined in the constructor (init) so that they become the part of the model's state. Defining them inside the constructor automatically register their parameters. If layers were definied inside the forward function new layer objects would be instantiated on each call, and PyTorch would not properly track their parameters across iterations. Activation functions like ReLU() can be defined in the forward() function but learnable layers must reside in init() function for the optimizer to recognize and update them.**
