## Q1: One-Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder

words = [["cat"], ["dog"], ["bird"], ["cat"]]

encoder = OneHotEncoder()

encoded = encoder.fit_transform(words)

print(encoded.toarray())
```

## Q2: TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "I love NLP",
    "NLP is amazing",
    "Machine learning is fun"
]

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(documents)

print(tfidf_matrix.toarray())
```

## Q3: CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

docs = ["apple orange apple","orange banana"]

vectorizer = CountVectorizer()

matrix = vectorizer.fit_transform(docs)

print(matrix.toarray())
```

## Q11: Stopword Removal

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

words = ["I", "love", "to", "learn", "NLP"]

stop_words = set(stopwords.words("english"))

filtered = [word for word in words if word.lower() not in stop_words]

print(filtered)
```

## Q12: Stemming

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

words = ["playing", "played", "plays"]

stems = [stemmer.stem(word) for word in words]

print(stems)
```

## Q13: Lemmatization

```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

words = ["running", "better", "cars"]

lemmas = [lemmatizer.lemmatize(word, pos='v' if word=="running" else 'a' if word=="better" else 'n') for word in words]

print(lemmas)
```

## Q14: Remove URLs

```python
import re

text = "Visit https://google.com for more information."

clean = re.sub(r'https?://\S+|www\.\S+','',text)

print(clean.strip())
```

## Q15: Remove HTML

```python
import re

html = "<h1>Welcome</h1><p>This is NLP.</p>"

def remove_html_tags(text):
    pattern = re.compile(r"<.*?>")
    text = pattern.sub(" ",text)
    text = re.sub(r"\s+"," ",text)
    return text.strip()

clean = remove_html_tags(html)
print(clean)

```

## Q16: Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer

corpus = [
    "I love NLP",
    "NLP is fun"
]

#apply bag of words
cv = CountVectorizer()

X = cv.fit_transform(corpus)

print(cv.vocabulary_)

print(X.toarray())
```

## Q20: Complete NLP Preprocessing

```python
import re

text = "Hello!!! Visit https://abc.com Email: student@gmail.com 123"

text = text.lower()

text = re.sub(r"https?://\S+", "", text)
text = re.sub(r"\S+@\S+", "", text)
text = re.sub(r"\d+", "", text)
text = re.sub(r"[^\w\s]", "", text)
text = re.sub(r"\s+", " ", text).strip()

print(text)
```
