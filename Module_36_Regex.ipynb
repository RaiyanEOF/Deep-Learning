# Character Set []

```python
import re
```

```python
txt = "The rain spain"
x = re.findall(r"[a-b]",txt)
print(x)
```

# Escape Sequence

```python
txt = " Hi This is Akash, My age is 25"
x = re.findall(r"\d+",txt)
print(x)
```

# Digit character class

```python
txt = "that will be 59 dollars"
x = re.findall(r"\d",txt)
print(x)
```

```python
txt = "that will be 59 dollars"
x = re.findall(r"\d+",txt)
print(x)
```

```python
#match 2 ,  3 digits

txt = "that will be 59 dollars"
x = re.findall(r"\d{2}",txt)
print(x)
```

```python
#match 2 or 3 digits

txt = "that will be 592 43 1 dollars"
x = re.findall(r"\d{2,3}",txt)
print(x)
```

```python
txt = "this is my number 01905-672239"
x = re.findall(r"\d{5}-\d{6}",txt)
print(x)
```

```python
#at the beginning

txt = "123  123 this is my number 01905-672239"
x = re.findall(r"^\d+",txt)

print(x)
```

```python
# at the end
#at the beginning

txt = "123  123 this is my number 01905-672239"
x = re.findall(r"\d+$",txt)

print(x)
```

# Wildcard character .

```python
txt = "hello"
x = re.findall(r".",txt)
print(x)
```

```python
txt = "hello"
x = re.findall(r"he..o",txt)
print(x)
```

# Asterics

```python
txt = "akash@gmail.com"
x = re.findall(r".*@.*\..*",txt)

print(x)
```

```python
txt = "akash@gmail.com  bat@gmail.com cat@gmail.com mat@gmail.com dog@yahoo.com"
x = re.findall(r".*@.*\.com$",txt)
print(x)
```

# Plus Quantifier

```python
txt = "heo 12"
x = re.findall(r"\d+",txt)
print(x)
```

# Question mark

```python
txt = "hello helo"

x = re.findall(r"hel?o",txt)

print(x)
```

```python
txt = "heo hello he1o"

x = re.findall(r"he.?o",txt)

print(x)
```

# Capture and group

```python
txt = "ababab"
x = re.findall("ab+",txt)
print(x)
```

```python
#email

txt = "abc@gmail.com"

x = re.findall(r"(\w+)@(\w+)\.com",txt)

print(x)

for first,second in x :
  print(first,second)
```

```python
#name ,age

txt = "john 25, Alice 30, Bob 18"

x =re.findall(r"(\w+)\s(\d+)",txt)
print(x)

for name,age in x:
  print(name,age)
```

# Functions


## Findall

```python
txt = "My lucky numbers are 7,25,99"

x = re.findall(r"\d+",txt)
print(x)
```

# Search

```python
txt = "My lucky number is 25 19 22 1"
x = re.search(r"\d+",txt)
# print(x)

# print(x.start())
# print(x.end())
print(x.group())
```

```python
# email
txt = "contact us at support@phitron.io , you can also email us admin@yahoo.com"

match = re.search(r"\w+@\w+\.\w+",txt)

print(match.group())
```

## Split

```python
txt = "apple, banana, orange"

x = re.split(r",",txt)
print(x)
```

## Sub

```python
txt = "I have 2 cats and 3 dogs"
x = re.sub(r"\d+","#",txt)

print(x)
```

## Compile

```python
txt = "I have 2 cats and 15 dogs"
```

```python
pattern = re.compile(r"\d+")
```

```python
#1: compile + findall()
x = pattern.findall(txt)
print(x)
```

```python
#2 compile + search
x = pattern.search(txt)
print(x.group())
```

```python
#Compile + sub

x = pattern.sub("#",txt)
print(x)
```

```python
# Compile + split

pattern = re.compile(r",")

txt = "apple, banana, orange"

x = pattern.split(txt)

print(x)
```
