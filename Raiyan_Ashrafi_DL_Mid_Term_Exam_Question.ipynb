# Question 02: [ Marks 20 ]

Manual Gradient Calculation

Given:

z = w*x + b <br>
y_pred = sigmoid(z) <br>
L = −[ y log(y_pred) + (1−y) log(1−y_pred) ] <br>

a) Write dL/dy_pred <br>
b) Write dy_pred/dz <br>
c) Show that dL/dz = y_pred − y <br>

## Write Answer 02:
z = w*x + b
y_pred = sigmoid(z)
L = −[ y log(y_pred) + (1 − y) log(1 − y_pred) ]

a) Derivative of loss with respect to y_pred:

dL/dy_pred = (1 − y) / (1 − y_pred) − y / y_pred

b) Derivative of sigmoid:

Since y_pred = sigmoid(z),

dy_pred/dz = y_pred (1 − y_pred)

c) Using chain rule:

dL/dz = (dL/dy_pred) × (dy_pred/dz)

After simplifying, the terms cancel out and we get:

dL/dz = y_pred − y

This is the final important result.

# Question 04: [ Marks 15 ]

Create a scalar tensor x = 3.0 with gradient tracking enabled. <br>
Compute y = x² + 4x. <br>
Perform backpropagation. <br>
Print the gradient of x. <br>

Write the full PyTorch code yourself. <br>

## Write Answer 04:

```python
import torch

x = torch.tensor(3.0, requires_grad=True)

y = x**2 + 4*x

y.backward()

print("Value of x:", x.item())
print("Value of y:", y.item())
print("Gradient of x:", x.grad.item())
```

# Question 05: [ Marks 10 ]

Create two tensors:
w = 2.0 (requires_grad=True) <br>
b = 1.0 (requires_grad=True) <br>

Let input x = 3.0. <br>
Compute z = w*x + b. <br>
Use Sigmoid to get y_pred.
 <br>

Define Binary Cross Entropy manually (without built-in loss). <br>
Compute gradients using backward(). <br>
Print gradients of w and b. <br>

Write full code.

## Write Answer 05:

```python
import torch

w = torch.tensor(2.0, requires_grad=True)
b = torch.tensor(1.0, requires_grad=True)

x = torch.tensor(3.0)
y_true = torch.tensor(1.0)

z = w * x + b

y_pred = torch.sigmoid(z)

loss = -(y_true * torch.log(y_pred) + (1 - y_true) * torch.log(1 - y_pred))

loss.backward()

print("Prediction:", y_pred.item())
print("Loss:", loss.item())
print("Gradient of w:", w.grad.item())
print("Gradient of b:", b.grad.item())
```

# Question 07: [ Marks 20 ]


Create tensor w = 2.0 (requires_grad=True). <br>

Compute two different losses sequentially and call backward() twice. <br>
Print w.grad. <br>

Do not clear gradients. <br>
Explain what you observe. <br>

Write full code.


## Write Answer 07:



```python
import torch

w = torch.tensor(2.0, requires_grad=True)


loss1 = w**2
loss1.backward()

print("Gradient after first backward:", w.grad.item())

loss2 = 3*w
loss2.backward()
print("Gradient after second backward:", w.grad.item())
```

In this code, we created a tensor w = 2.0 with gradient tracking enabled. First, we calculated the loss as w^2 and called backward(). The derivative of w^2 is 2w, so when w = 2, the gradient becomes 4.

Next, we calculated a second loss as 3w and called backward() again without clearing the previous gradient. The derivative of 3w is 3. Instead of replacing the old gradient, PyTorch added this new gradient to the existing one.

So the final value of w.grad became 4 + 3 = 7.

This shows that PyTorch accumulates gradients by default. If we do not clear the gradients before calling backward() again, the gradients will keep adding together.
